```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./processed_data/filtered_data_with_classes.csv')
print(df.head(1))
```

```python
print(df.describe())
print(df.info())
```

```python
grouped = df.groupby('condition_text')

# Step 2: Calculate the mean for each group
group_means = grouped.mean()

# Step 3: Plot the averages for each class
group_means.plot(kind='bar', figsize=(10, 6), legend=True)
plt.title('Average of Columns by Condition Text')
plt.ylabel('Average Value')
plt.xlabel('Condition Text')
plt.xticks(rotation=45)  # Rotate labels to make them readable
plt.show()
```
